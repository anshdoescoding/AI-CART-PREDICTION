### Project Overview
This project builds a product recommendation system using customer cart data. Product names are cleaned, and a co-occurrence matrix is created to find items frequently bought together. Based on this, the system predicts the next likely product and recommends three additional items. Regression models are then applied to analyze product popularity, where Polynomial Regression (degree 3) shows the best trend fit. The project effectively combines recommendation logic with demand analysis.

## TRAINING THE MODEL

```python
def normalize_name(name: str) -> str:
    if pd.isna(name):
        return ""
    name = str(name).lower()
    name = re.sub(r"[^a-z0-9\s]", " ", name)
    name = re.sub(r"\b(ml|g|gm|kg|pcs|piece|pack|ltr|liter|litre|unit|bottle|pouch|combo|set)\b", "", name)
    name = re.sub(r"\s+", " ", name).strip()
    return name


def build_cooccurrence_matrix(df, cart_col):
    cooccur = defaultdict(Counter)

    for _, row in df.iterrows():
        cart = str(row[cart_col]).split(",")
        items = [normalize_name(x.strip()) for x in cart if x.strip()]
        for i in range(len(items)):
            for j in range(i + 1, len(items)):
                a, b = items[i], items[j]
                cooccur[a][b] += 1
                cooccur[b][a] += 1
    return cooccur


def predict_products(cart_items, cooccur, all_products, n=1):
    scores = Counter()
    cart_norm = [normalize_name(x) for x in cart_items]
    for item in cart_norm:
        for other, count in cooccur.get(item, {}).items():
            if other not in cart_norm:
                scores[other] += count
    if not scores:
        return all_products[:n]
    return [p for p, _ in scores.most_common(n)]


def recommend_products(cart_items, cooccur, all_products, exclude=None, n=3):
    exclude = set(normalize_name(x) for x in (exclude or []))
    cart_norm = set(normalize_name(x) for x in cart_items)
    scores = Counter()

    for item in cart_norm:
        for other, count in cooccur.get(item, {}).items():
            if other not in cart_norm and other not in exclude:
                scores[other] += count

    recs = [p for p, _ in scores.most_common(n)]
    if len(recs) < n:
        extra = [p for p in all_products if p not in cart_norm and p not in exclude]
        recs.extend(extra[: n - len(recs)])
    return recs
```

## PRODUCT-TO-MRP MAP

```python
# Step 4: Build a more reliable product-to-MRP map
if 'product' in df.columns and 'mrp' in df.columns:
    product_df = df[['product', 'mrp']].dropna().copy()
else:
    print("‚ö†Ô∏è 'product' and 'mrp' columns not found, searching automatically...")
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    text_cols = df.select_dtypes(include=['object']).columns.tolist()
    print("Numeric columns:", num_cols)
    print("Text columns:", text_cols)
    product_df = pd.DataFrame(columns=['product', 'mrp'])

product_df['product_norm'] = product_df['product'].apply(normalize_name)
mrp_map = product_df.groupby('product_norm')['mrp'].mean().to_dict()
all_products = list(mrp_map.keys())

print(f"‚úÖ Catalog built with {len(all_products)} products.")
```

## GENERATING OUTPUT

```python
cart_col = "Products in cart"   # üëà change if your column name is different

if cart_col not in df.columns:
    print("‚ö†Ô∏è Column", cart_col, "not found. Available columns are:", df.columns.tolist())
else:
    cooccur = build_cooccurrence_matrix(df, cart_col)

    df["Predicted Product"] = ""
    df["Recommended Products"] = ""

    for idx, row in df.iterrows():
        cart = str(row[cart_col]).split(",")
        pred = predict_products(cart, cooccur, all_products, n=1)
        recs = recommend_products(cart, cooccur, all_products, exclude=pred, n=3)

        df.at[idx, "Predicted Product"] = ", ".join(pred)
        df.at[idx, "Recommended Products"] = ", ".join(recs)

    print("‚úÖ Predictions and recommendations generated successfully!\n")

    display_columns = [cart_col, "Predicted Product", "Recommended Products"]
    display(df[display_columns])
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# --- Reuse your frequency dataframe from earlier ---
X = freq_df[["Product_Index"]]
y = freq_df["Count"]

# -------------------- MLR --------------------
X_mlr = pd.DataFrame({
    "Product_Index": freq_df["Product_Index"],
    "Log_Index": np.log1p(freq_df["Product_Index"])
})

mlr = LinearRegression()
mlr.fit(X_mlr, y)
y_pred_mlr = mlr.predict(X_mlr)

r2_mlr = r2_score(y, y_pred_mlr)
rmse_mlr = np.sqrt(mean_squared_error(y, y_pred_mlr))
mae_mlr = mean_absolute_error(y, y_pred_mlr)

print("üìó Multiple Linear Regression (MLR) Results")
print(f"R¬≤ Score: {r2_mlr:.4f}")
print(f"RMSE: {rmse_mlr:.4f}")
print(f"MAE: {mae_mlr:.4f}\n")

# -------------------- PLR (degree 2) --------------------
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)
plr2 = LinearRegression().fit(X_poly2, y)
y_pred_plr2 = plr2.predict(X_poly2)

r2_plr2 = r2_score(y, y_pred_plr2)
rmse_plr2 = np.sqrt(mean_squared_error(y, y_pred_plr2))
mae_plr2 = mean_absolute_error(y, y_pred_plr2)

# -------------------- PLR (degree 3) --------------------
poly3 = PolynomialFeatures(degree=3)
X_poly3 = poly3.fit_transform(X)
plr3 = LinearRegression().fit(X_poly3, y)
y_pred_plr3 = plr3.predict(X_poly3)

r2_plr3 = r2_score(y, y_pred_plr3)
rmse_plr3 = np.sqrt(mean_squared_error(y, y_pred_plr3))
mae_plr3 = mean_absolute_error(y, y_pred_plr3)

print("üìô Polynomial Regression (Degree 3) Results")
print(f"R¬≤ Score: {r2_plr3:.4f}")
print(f"RMSE: {rmse_plr3:.4f}")
print(f"MAE: {mae_plr3:.4f}")

# -------------------- Visualization --------------------
plt.figure(figsize=(8,6))
plt.scatter(X, y, color="orange", label="Actual Data")

X_seq = np.linspace(X.min(), X.max(), 300).reshape(-1,1)
plt.plot(X_seq, plr2.predict(poly2.transform(X_seq)), color="blue", linewidth=2, label="PLR Degree 2")
plt.plot(X_seq, plr3.predict(poly3.transform(X_seq)), color="red", linewidth=2, label="PLR Degree 3")

plt.title("Polynomial Regression Comparison (Degree 2 vs 3)")
plt.xlabel("Product Index")
plt.ylabel("Count")
plt.legend()
plt.show()

# -------------------- Summary Table --------------------
results_ext = pd.DataFrame({
    "Model": ["MLR", "PLR (Degree 2)", "PLR (Degree 3)"],
    "R¬≤ Score": [r2_mlr, r2_plr2, r2_plr3],
    "RMSE": [rmse_mlr, rmse_plr2, rmse_plr3],
    "MAE": [mae_mlr, mae_plr2, mae_plr3]
})
display(results_ext)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Use your frequency dataframe
X = freq_df[["Product_Index"]]
y = freq_df["Count"]

# ---------------- SLR ----------------
slr = LinearRegression().fit(X, y)
y_pred_slr = slr.predict(X)
r2_slr = r2_score(y, y_pred_slr)
rmse_slr = np.sqrt(mean_squared_error(y, y_pred_slr))

# ---------------- MLR ----------------
X_mlr = pd.DataFrame({
    "Product_Index": freq_df["Product_Index"],
    "Log_Index": np.log1p(freq_df["Product_Index"])
})
mlr = LinearRegression().fit(X_mlr, y)
y_pred_mlr = mlr.predict(X_mlr)
r2_mlr = r2_score(y, y_pred_mlr)
rmse_mlr = np.sqrt(mean_squared_error(y, y_pred_mlr))

# ---------------- PLR Degree 2 ----------------
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)
plr2 = LinearRegression().fit(X_poly2, y)
y_pred_plr2 = plr2.predict(X_poly2)
r2_plr2 = r2_score(y, y_pred_plr2)
rmse_plr2 = np.sqrt(mean_squared_error(y, y_pred_plr2))

# ---------------- PLR Degree 3 ----------------
poly3 = PolynomialFeatures(degree=3)
X_poly3 = poly3.fit_transform(X)
plr3 = LinearRegression().fit(X_poly3, y)
y_pred_plr3 = plr3.predict(X_poly3)
r2_plr3 = r2_score(y, y_pred_plr3)
rmse_plr3 = np.sqrt(mean_squared_error(y, y_pred_plr3))

# ---------------- Ridge Regression ----------------
ridge = Ridge(alpha=1.0).fit(X_poly3, y)
y_pred_ridge = ridge.predict(X_poly3)
r2_ridge = r2_score(y, y_pred_ridge)
rmse_ridge = np.sqrt(mean_squared_error(y, y_pred_ridge))

# ---------------- Lasso Regression ----------------
lasso = Lasso(alpha=0.001, max_iter=10000).fit(X_poly3, y)
y_pred_lasso = lasso.predict(X_poly3)
r2_lasso = r2_score(y, y_pred_lasso)
rmse_lasso = np.sqrt(mean_squared_error(y, y_pred_lasso))

# ---------------- Combine Results ----------------
results = pd.DataFrame({
    "Model": ["SLR", "MLR", "PLR (2)", "PLR (3)", "Ridge", "Lasso"],
    "R¬≤": [r2_slr, r2_mlr, r2_plr2, r2_plr3, r2_ridge, r2_lasso],
    "RMSE": [rmse_slr, rmse_mlr, rmse_plr2, rmse_plr3, rmse_ridge, rmse_lasso]
})

print("üìä Model Performance Comparison:\n")
display(results)

# ---------------- Visualization ----------------
plt.figure(figsize=(8,5))
sns.barplot(data=results, x="Model", y="R¬≤", palette="viridis")
plt.title("Model R¬≤ Comparison")
plt.ylabel("R¬≤ Score")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=results, x="Model", y="RMSE", palette="magma")
plt.title("Model RMSE Comparison")
plt.ylabel("RMSE")
plt.show()
```
